```python
from os import truncate
from langchain_community.document_loaders import YoutubeLoader

video_url = "https://www.youtube.com/watch?v=31qyMKNB2RA&t=4144s"

loader = YoutubeLoader.from_youtube_url(
    video_url,
    language=["hi"],

)

docs = loader.load()

print("Transcript loaded ✅")
print(docs[0].page_content[500:])
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=100
)

chunks = text_splitter.split_documents(docs)

print(f"Total chunks created: {len(chunks)}")
print(chunks[0].page_content[:300])
```

```python
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings
import os

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)


faiss_index_path = "yt_faiss_index"
faiss_index_file = os.path.join(faiss_index_path, "index.faiss")

if os.path.exists(faiss_index_file):
    print("Loading existing FAISS index...")
    vectorstore = FAISS.load_local(
        faiss_index_path,
        embeddings,
        allow_dangerous_deserialization=True
    )
    print("FAISS index loaded ✅")
elif 'chunks' in locals() and 'embeddings' in locals(): # Check if chunks are available
    print("FAISS index not found. Creating a new one...")
    vectorstore = FAISS.from_documents(
        chunks,
        embeddings
    )
    vectorstore.save_local(faiss_index_path)
    print("FAISS index created and saved ✅")
else:
    print("Error: Chunks or embeddings not available to create FAISS index.")
    print("Please ensure previous steps (document loading and chunking) are executed.")
```

```python
from langchain_community.vectorstores import FAISS

vectorstore = FAISS.from_documents(
    chunks,
    embeddings
)

print("FAISS index created ✅")
```

```python
vectorstore.index_to_docstore_id
```

```python
retriever = vectorstore.as_retriever(
    search_type="mmr",
    search_kwargs={
        "k": 10,
        "fetch_k": 20,
        "lambda_mult": 0.6
    }
)
```
